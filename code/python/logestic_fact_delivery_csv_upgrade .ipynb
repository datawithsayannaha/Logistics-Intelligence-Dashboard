```python
import pandas as pd
```

```python
fact = pd.read_csv("fact_delivery.csv")
```

```python
print(fact.shape)
```

```python
print(fact.columns)
```

```python
fact['purchase_date'] = pd.to_datetime(fact['purchase_date'])
fact['delivered_date'] = pd.to_datetime(fact['delivered_date'])
fact['estimated_date'] = pd.to_datetime(fact['estimated_date'])
```

```python
fact["route_name"] = fact["seller_state"] + " -> " + fact["customer_state"]
```

```python
fact[["seller_state", "customer_state", "route_name"]].head(10)
```

```python
def delivery_bucket(x):
    if pd.isna(x):
        return "Not Delivered"
    elif x < 0:
        return "Early"
    elif x == 0:
        return "On Time"
    elif x <= 3:
        return "Slight Delay"
    else:
        return "High Delay"

fact["delivery_status_bucket"] = fact["delay_days"].apply(delivery_bucket)
```

```python
fact["delivery_status_bucket"].head(10)
```

```python
fact["volume_cm3"] = (
    fact["product_length_cm"] *
    fact["product_height_cm"] *
    fact["product_width_cm"]
)
```

```python
fact["volume_cm3"].head()
```

```python
def weight_bucket(x):
    if pd.isna(x):
        return "Unknown"
    elif x <= 500:
        return "Light"
    elif x <= 2000:
        return "Medium"
    else:
        return "Heavy"

fact["weight_bucket"] = fact["product_weight_g"].apply(weight_bucket)
```

```python
fact["weight_bucket"].head(10)
```

```python
def size_bucket(x):
    if pd.isna(x):
        return "Unknown"
    elif x <= 1000:
        return "Small"
    elif x <= 5000:
        return "Medium"
    else:
        return "Large"

fact["size_bucket"] = fact["volume_cm3"].apply(size_bucket)
```

```python
fact["size_bucket"].head(10)
```

```python
fact.head()
```

```python
print(fact.columns)
```

```python
print(fact.shape)
```

```python
fact["route_type"] = fact.apply(
    lambda x: "Intra-State" if x["seller_state"] == x["customer_state"] else "Inter-State",
    axis=1
)
```

```python
fact["route_type"].head(100)
```

```python
def delivery_speed(x):
    if pd.isna(x):
        return "Not Delivered"
    elif x <= 3:
        return "Fast"
    elif x <= 7:
        return "Standard"
    else:
        return "Slow"

fact["delivery_speed"] = fact["delivery_days"].apply(delivery_speed)
```

```python
fact["delivery_speed"].head(10)
```

```python
def freight_bucket(x):
    if pd.isna(x):
        return "Unknown"
    elif x <= 20:
        return "Low Cost"
    elif x <= 50:
        return "Medium Cost"
    else:
        return "High Cost"

fact["freight_bucket"] = fact["freight_value"].apply(freight_bucket)
```

```python
fact["freight_bucket"].head(10)
```

```python
def order_value_bucket(x):
    if pd.isna(x):
        return "Unknown"
    elif x <= 50:
        return "Low Value"
    elif x <= 150:
        return "Medium Value"
    else:
        return "High Value"

fact["order_value_bucket"] = fact["price"].apply(order_value_bucket)
```

```python
fact["order_value_bucket"].head(10)
```

```python
fact["package_density"] = fact["product_weight_g"] / fact["volume_cm3"]
```

```python
def density_bucket(x):
    if pd.isna(x):
        return "Unknown"
    elif x <= 0.05:
        return "Low Density"
    elif x <= 0.20:
        return "Medium Density"
    else:
        return "High Density"

fact["density_bucket"] = fact["package_density"].apply(density_bucket)
```

```python
fact["density_bucket"].head(10)
```

```python
fact["delivery_risk_flag"] = fact.apply(
    lambda x: 1 if (x["delayed_flag"] == 1 and x["product_weight_g"] > 2000 and x["freight_value"] > 50) else 0,
    axis=1
)
```

```python
fact["delivery_risk_flag"].head(10)
```

```python
fact["purchase_month"] = fact["purchase_date"].dt.month_name()
fact["purchase_year"] = fact["purchase_date"].dt.year
fact["purchase_quarter"] = "Q" + fact["purchase_date"].dt.quarter.astype(str)
```

```python
fact.head()
```

```python
print(fact.columns)
```

```python
fact["purchase_month_num"] = fact["purchase_date"].dt.month
```

```python
fact.to_csv("fact_delivery_enriched.csv", index=False)
```
